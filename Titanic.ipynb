Median was used to replace null values for Age as it had the most central point

```python
df.head(20)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Cabin_numeric'] = le.fit_transform(df['Cabin'])
correlation=df['Cabin_numeric'].corr(df['Fare'])
print(correlation)
```

```python
fare_group=df.groupby('Cabin')['Fare'].mean()
print (fare_group)
```

Mean was used to fill in blank cells for fare, because it was the most central point.

```python
mean_fare=df.loc[df['Cabin'].isnull(), 'Fare'].mean()
print(mean_fare)
```

##Data Analysis

```python
group_survived=df.groupby('Cabin')['Survived'].sum().sort_values(ascending = False)
group_survived.head(30)
```

```python
df.head()
```

```python
sex_survived=df.groupby('Sex')['Survived'].sum()
sex_survived.head()
```

```python
sex_survived.plot(kind = 'bar', x = 'Sex', y='Survived')
plt.show()
```

```python
embarked_survived=df.groupby('Embarked')['Survived'].sum()
embarked_survived.head()
```

```python
embarked_survived.plot(kind = 'bar', x = 'Embarked', y='Survived')
plt.show()
```

```python
df[['Survived','Parch']].corr()
```

```python
import seaborn as sns
corr1=df[['Survived','Parch']].corr()
sns.heatmap(corr1, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
```

```python
class_survived=df.groupby('Pclass')['Survived'].sum()
class_survived.head()
```

```python
class_survived.plot(kind = 'bar', x = 'Pclass', y='Survived')
plt.show()
```

# Findings


*   The 3rd class had the highest survival count as they were prioritize during the evacuation.
*   The 2nd class was highly affected because they evacuation procedure was not properly done, and they were in the innermost part of the ship.
*   Southampton had the highest survival count. This could be because most passengers disembarked on the following ports.
*   There were no male survivors compared to women since women and children were given priority when evacuating.

# Recommendations

*   A proper evacuation procedure to be put in place.
*   Ensure all passenger classes have uniformed evacuation points in case of emergencies.
*   Ensure the ship staff/ personnel are well trained and follow the evacuation policy.
